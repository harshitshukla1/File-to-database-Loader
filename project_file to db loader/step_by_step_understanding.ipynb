```python
import pandas as pd 
# the CSV’s first row is actual data, not column names, so read without a
# header and then assign sensible names
df = pd.read_csv('sample.csv', header=None)
df.columns = ['id', 'username', 'age', 'email']

# now the dataframe has valid identifiers and the INSERT will succeed
df.to_sql('users', con=engine, if_exists='append', index=False)
```

```python
#create table via notebook
%%sql
create table if not exists employee(
    id int primary key,
    name varchar(100),
    age int,
    email varchar(50)

);
```

```python
#insert value via notebook by making dataframe through python 
new_df=pd.DataFrame({
    'id':[1,2],
    'name':['John','Rita'],
    'age':[40,35],
    'email':['john@gmail.com','Rita@gmail.com']
})
new_df.to_sql('employee', con= engine, if_exists='append',index=False)
```

```python
#for reading the large dataset into small chunks
chunk_size=3
for chunk in pd.read_csv('sample.csv', chunksize=chunk_size, header=None):
    chunk.columns = ['id', 'username', 'age', 'email']
    chunk.to_sql('users', con=engine, if_exists='append', index=False)
```

```python
#load all available .csv file in a folder into databae automatically
import os
data_dir='data/'
for file in os.listdir(data_dir):
    if file.endswith('.csv'):
        table_name=os.path.splitext(file)[0]
        print(f"Loading {file} into table {table_name}")
        df=pd.read_csv(os.path.join(data_dir,file))
        df.to_sql(table_name,con=engine,if_exists='replace',index=False)
```

```python
#Add logging for number of rows loaded per table:
for file in os.listdir(data_dir):
    if file.endswith('.csv'):
        table_name = os.path.splitext(file)[0]
        df = pd.read_csv(os.path.join(data_dir, file))
        df.to_sql(table_name, con=engine, if_exists='replace', index=False)
        print(f"Loaded {len(df)} rows into table {table_name}")
```
